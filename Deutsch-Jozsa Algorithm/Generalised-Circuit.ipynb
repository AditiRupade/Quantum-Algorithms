```python
from qiskit import IBMQ, Aer
from qiskit.providers.ibmq import least_busy
from qiskit import QuantumCircuit, assemble, transpile
import numpy as np
from qiskit.visualization import plot_histogram
```

### Oracle

```python
def oracle(oracle_type, n):
    
    # Input : oracle_type - string ("balanced"/"constant")
    # Input : n - integer (input register length)
    # Output : oracle_gate
    
    oracle_circuit = QuantumCircuit(n+1)
    
    if oracle_type.lower() == "balanced":
        # Generate random bits to wrap CNOTs in X gates
        random_bits = np.random.randint(1,2**n)
        bit_string = format(random_bits, '0'+str(n)+'b')
        # Apply X gate for 1 in bit_string
        for qubit in range(len(bit_string)):
            if bit_string[qubit] == '1':
                oracle_circuit.x(qubit)
        # Apply CNOT gate to generate random oracle
        for qubit in range(n):
            oracle_circuit.cx(qubit,n)
        # Apply X gate to redo previous operations
        for qubit in range(len(bit_string)):
            if bit_string[qubit] == '1':
                oracle_circuit.x(qubit)
                
    if oracle_type.lower() == "constant":
        # Generate 0's and 1's randomly to decide X operation
        output = np.random.randint(2)
        if output == 1:
            oracle_circuit.x(n)
            
    # Convert the circuit to gate
    oracle_gate = oracle_circuit.to_gate()
    oracle_gate.name = "Oracle"
    return oracle_gate
```

### Deutsch-Jozsa Algorithm

```python
def deutsch_jozsa_algorithm(oracle, n):
    # Input : oracle - gate that can be appended to a circuit as an oracle
    # Input : n - input register length
    # Output : circuit - deutsch-jozsa circuit with complete algorithm till measurement
    # If circuit measurement is 0 then oracle is constant. If it is 1, then oracle is balanaced.
    
    deutsch_jozsa_circuit = QuantumCircuit(n+1, n)
    
    # Step 1 : Apply X gate to second register
    deutsch_jozsa_circuit.x(n)
    
    # Step 2 : Apply Hadamard gate to each qubit
    for qubit in range(n+1):
        deutsch_jozsa_circuit.h(qubit)
        
    # Step 3 : Apply oracle
    deutsch_jozsa_circuit.append(oracle, range(n+1))
    
    # Step 4 : Apply Hadamard on input register
    for qubit in range(n):
        deutsch_jozsa_circuit.h(qubit)
        
    # Step 5 : Measure the input register
    for i in range(n):
        deutsch_jozsa_circuit.measure(i,i)
        
    return deutsch_jozsa_circuit
```

### Testing the Functions

```python
n = 4
balanced_oracle_gate = oracle('balanced', n)
balanaced_deutsch_jozsa_circuit = deutsch_jozsa_algorithm(balanced_oracle_gate, n)
balanaced_deutsch_jozsa_circuit.draw('mpl')
```

```python
# Output using Local Simulator
simulator = Aer.get_backend('aer_simulator')
transpiled_circuit = transpile(balanaced_deutsch_jozsa_circuit, simulator)
qobj = assemble(transpiled_circuit, simulator)
results = simulator.run(qobj).result()
answer = results.get_counts()

plot_histogram(answer)
```

```python
n = 4
constant_oracle_gate = oracle('constant', n)
constant_deutsch_jozsa_circuit = deutsch_jozsa_algorithm(constant_oracle_gate, n)
constant_deutsch_jozsa_circuit.draw('mpl')
```

```python
# Output using Local Simulator
simulator = Aer.get_backend('aer_simulator')
transpiled_circuit = transpile(constant_deutsch_jozsa_circuit, simulator)
qobj = assemble(transpiled_circuit, simulator)
results = simulator.run(qobj).result()
answer = results.get_counts()

plot_histogram(answer)
```
